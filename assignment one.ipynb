```python
import random

class MarsEnvironment:
    def __init__(self):
        self.locations = {
            "A": {"has_rocks": True, "sampled": False},
            "B": {"has_rocks": True, "sampled": False},
            "C": {"has_rocks": False, "sampled": False},
            "D": {"has_rocks": False, "sampled": False}
        }

    def update_environment(self):
        for location in self.locations.values():
            location["has_rocks"] = random.choice([True, False])

class RoverAgent:
    def __init__(self, environment):
        self.environment = environment
        self.current_location = random.choice(list(environment.locations.keys()))
        self.samples = []

    def explore(self, environment):
        location = self.environment.locations[self.current_location]
        if not location["sampled"]:
            location["sampled"] = True
            if location["has_rocks"]:
                self.samples.append(self.current_location)
                print(f"Rocks sampled at location {self.current_location}")
            else:
                print(f"No rocks found at location {self.current_location}")
        else:
            print(f"Location {self.current_location} has already been sampled")

        self.current_location = random.choice(list(environment.locations.keys()))

class RoverPerformance:
    def __init__(self, samples):
        self.samples = samples

    def calculate_performance(self):
        if self.samples:
            total_locations = len(self.samples)
            unique_locations = len(set(self.samples))
            performance = (unique_locations / total_locations) * 100
            print(f"Rover performance: {performance:.2f}%")
            
        
        else:
            print("No samples collected.cannot calculate perfomance.")
        
    

def main():
    environment = MarsEnvironment()
    rover_agent = RoverAgent(environment)
    rover_performance = RoverPerformance(rover_agent.samples)

    for _ in range(2):  
        print(f"\nExploration Venture {_ + 1}")
        environment.update_environment()
        rover_agent.explore(environment)
        rover_agent.explore(environment)
        print(f"\nLocations Sampled: {rover_agent.samples}")
        rover_performance.calculate_performance()

if __name__ == "__main__":
    main()
```
